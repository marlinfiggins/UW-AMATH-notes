```julia
using Plots
```

## Exercise 1

```julia
function uniform_exer_1(x, ϵ)
   A =  1 - exp(1/2)
    
    return exp(1/(x+1) - 1/2) + A*exp(- x/ϵ) + 1 - A - exp(1/2)
end
```

```julia
ϵ_vec = [0.01, 0.05, 0.1, 0.2]

exer_1 = plot(title = "Uniform Solution (Exercise 1)",
        xlabel = "x")
for ϵ in ϵ_vec
    exer_1 = plot!(x -> uniform_exer_1(x, ϵ), 0, 1,
                label = "ϵ = $ϵ")
end

exer_1
```

```julia
savefig(exer_1, "../hw/figs/hw-5-exer-1.png")
```

## Exercise 2

```julia
function uniform_exer_2(x, ϵ)
    return exp(- x / sqrt(ϵ)) + exp(-(1-x)/ϵ)
end
```

```julia
ϵ_vec = [0.01, 0.05, 0.1, 0.2]

exer_2 = plot(title = "Uniform Solution (Exercise 2)",
        xlabel = "x")
for ϵ in ϵ_vec
    exer_2 = plot!(x -> uniform_exer_2(x, ϵ), 0, 1,
                label = "ϵ = $ϵ")
end

exer_2
```

```julia
savefig(exer_2, "../hw/figs/hw-5-exer-2.png")
```
