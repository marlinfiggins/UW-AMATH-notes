```julia
using Plots
```

```julia
function uniform_exer_1(x, ϵ)

    u1 = exp(1/(x+1) - 1/2)
    u2 = ((1 - exp(1/2)) / (1 + x)^2) * exp(-1/(x+1) + 1 - ((1+x)^3  - 1)/(3ϵ) )
    return u1 + u2
end
```

```julia
ϵ_vec = [0.01, 0.05, 0.1, 0.2]

exer_1 = plot(title = "Uniform Solution (Exercise 1)",
        xlabel = "x")
for ϵ in ϵ_vec
    exer_1 = plot!(x -> uniform_exer_1(x, ϵ), 0, 1,
                label = "ϵ = $ϵ")
end

exer_1
```
